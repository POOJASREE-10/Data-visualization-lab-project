```python
# Creating SQLite database

import pandas as pd
import sqlite3
from google.colab import files

uploaded = files.upload()

filename = list(uploaded.keys())[0]
df = pd.read_excel(filename)

conn = sqlite3.connect("Food_Waste.db")

df.to_sql("Food_Waste_Prediction", conn, if_exists="replace", index=False)

conn.close()

print("SQLite database created successfully!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/POOJASREE-10/Data-visualization-lab-project/main/Food_Waste_Prediction_Dataset.csv")

# Count the number of records for each cuisine
cuisine_count = df['Cuisine'].value_counts()

plt.figure(figsize=(6,4))
plt.bar(cuisine_count.index, cuisine_count.values)
plt.title("Distribution of Cuisine Types")
plt.xlabel("Cuisine")
plt.ylabel("Count")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/POOJASREE-10/Data-visualization-lab-project/main/Food_Waste_Prediction_Dataset.csv")

# Count the top 10 food items
food_count = df['Food_Item'].value_counts().head(10)

plt.figure(figsize=(10,5))
plt.bar(food_count.index, food_count.values)
plt.title("Top 10 Food Items")
plt.xlabel("Food Item")
plt.ylabel("Number of Records")
plt.xticks(rotation=45)
plt.show()
```
