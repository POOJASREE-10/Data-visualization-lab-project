```python
# ==========================================
# CRISP-DM - Food Waste Prediction Dataset
# ==========================================

import pandas as pd
from google.colab import files
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# -----------------------------------
# Phase 1: Business Understanding
# Objective: Predict High Food Waste
# -----------------------------------

# Upload Dataset
uploaded = files.upload()

# Read Dataset
df = pd.read_csv("Food_Waste_Prediction_Dataset.csv")

# -----------------------------------
# Phase 2: Data Understanding
# -----------------------------------

print("First 5 Records")
print(df.head())

print("\nDataset Information")
print(df.info())

print("\nMissing Values")
print(df.isnull().sum())

print("\nStatistical Summary")
print(df.describe())

# -----------------------------------
# Phase 3: Data Preparation
# -----------------------------------

# Convert Date column to datetime
df["Date"] = pd.to_datetime(df["Date"])

# Extract useful features
df["Month"] = df["Date"].dt.month
df["Day"] = df["Date"].dt.day
df["Weekday"] = df["Date"].dt.weekday

# Remove Date column
df.drop("Date", axis=1, inplace=True)

# Create Target Variable
median = df["Waste_Amount_kg"].median()
df["High_Waste"] = (df["Waste_Amount_kg"] > median).astype(int)

# Remove original Waste column
df.drop("Waste_Amount_kg", axis=1, inplace=True)

# Fill Missing Values
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].fillna(df[col].mode()[0])
    else:
        df[col] = df[col].fillna(df[col].median())

# Encode Categorical Columns
encoder = LabelEncoder()

for col in df.select_dtypes(include="object").columns:
    df[col] = encoder.fit_transform(df[col])

# Separate Features and Target
X = df.drop("High_Waste", axis=1)
y = df["High_Waste"]

# Handle Missing Values
imputer = SimpleImputer(strategy="mean")
X = imputer.fit_transform(X)

# Feature Scaling
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

# -----------------------------------
# Phase 4: Modeling
# -----------------------------------

model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

model.fit(X_train, y_train)

# -----------------------------------
# Phase 5: Evaluation
# -----------------------------------

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print("\n==============================")
print("CRISP-DM Accuracy:", round(accuracy * 100, 2), "%")
print("==============================")

# -----------------------------------
# Phase 6: Deployment
# -----------------------------------

# import joblib
# joblib.dump(model, "FoodWastePredictionModel.pkl")
```
